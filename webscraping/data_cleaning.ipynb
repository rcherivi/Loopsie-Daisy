```python
import pandas as pd
import os
```

```python
all_free_df = pd.read_csv("../data/all_free_crochet_patterns.csv")
cols = ['title', 'description', 'skill_level', 'Crochet Hook', 'Yarn Weight', 'Finished Size']

all_free_df['final_description'] = (
    all_free_df[cols]
    .fillna('')
    .agg(' '.join, axis=1)
    .str.replace(r'\s+', ' ', regex=True)
    .str.strip()
)
all_free_df.head()

all_free_df.rename(columns={'Crochet Hook': 'crochet_hook', 
'Yarn Weight': 'yarn_weight',
'Crochet Gauge': 'crochet_gauge',
'Finished Size': 'finished_size'}, inplace=True)
all_free_df.head()
# print(all_free_df.iloc[0,1])
```

```python
crochet_df = pd.read_csv("../data/crochet_crochet_patterns.csv")

# dropping this, has all nans
crochet_df = crochet_df.drop(columns=['Yarn Line', 'Yardage'])
description_col = crochet_df.columns[:-3]
print(description_col)

crochet_df['final_description'] = (
    crochet_df[description_col]
    .fillna('')
    .agg(' '.join, axis=1)
    .str.replace(r'\s+', ' ', regex=True)
    .str.strip()
)

crochet_df.rename(columns = {'Fiber Type': 'fiber_type', 'Pattern Type': 'pattern_type', 'Yarn Weight': 'yarn_weight', 
'Sizes Included': 'finished_size', 'Difficulty': 'skill_level', 'Needle / Hook Sizes': 'crochet_hook'}, 
inplace = True)

crochet_df.head()
```

```python
herrschners_df = pd.read_csv("../data/herrschners_crochet_patterns.csv")
herrschners_df = herrschners_df.drop(columns=['sku','upc'])
herrschners_df.head()
herrschners_df.rename(columns = {'product_link': 'pattern_link', 'skill': 'pattern_type', 
'difficulty': 'skill_level'}, inplace = True)
exclude_cols = ['pattern_link', 'review_count', 'image_url', 'local_path', 'star_rating', 'thread_size']
description_col = [col for col in herrschners_df.columns if col not in exclude_cols]
print(description_col)

herrschners_df['final_description'] = (
    herrschners_df[description_col]
    .fillna('')
    .agg(' '.join, axis=1)
    .str.replace(r'\s+', ' ', regex=True)
    .str.strip()
)
herrschners_df.head()
```

```python
lovecrafts_df = pd.read_csv("../data/lovecrafts_crochet_patterns.csv")
lovecrafts_df = lovecrafts_df.drop(columns = ['format', 'merged_info',
'techniques_and_construction', 'needles_required', 'crochet_thread_size'])
lovecrafts_df.head()

lovecrafts_df.rename(columns = {'product_link': 'pattern_link', 'craft': 'pattern_type'}, inplace = True)
exclude_cols = ['pattern_link', 'image_url', 'local_path']

lovecrafts_df['pages'] = lovecrafts_df['pages'].astype(str) + 'pages'
lovecrafts_df['number_of_patterns'] = lovecrafts_df['number_of_patterns'].astype(str) + 'patterns'

description_col = [col for col in lovecrafts_df.columns if col not in exclude_cols]
print(description_col)

lovecrafts_df['final_description'] = (
    lovecrafts_df[description_col]
    .fillna('')
    .agg(' '.join, axis=1)
    .str.replace(r'\s+', ' ', regex=True)
    .str.strip()
)
lovecrafts_df.head()
```

```python
herrschners_df['description'] = herrschners_df['description'].str.split("Skill Level").str[0].str.strip()
```

```python
lovecrafts_df['title'] = lovecrafts_df['title'].str.split(" - Downloadable PDF").str[0].str.strip()
```

```python
features = ['title', 'description', 'final_description', 'skill_level', 'pattern_link', 'local_path']
```

```python
combined_df = pd.concat(
    [
        all_free_df[features],
        crochet_df[features],
        lovecrafts_df[features],
        herrschners_df[features]
    ],
    axis=0,
    ignore_index=True
)

combined_df.shape
```

```python
combined_df = combined_df[~combined_df[["title", "pattern_link", "local_path"]].isna().any(axis=1)]
```

```python
combined_df["local_path"] = combined_df["local_path"].str.replace(
    r"\.\./data/images[\\/]", "", regex=True
)
```
